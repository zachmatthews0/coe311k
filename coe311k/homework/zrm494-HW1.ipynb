```python
import numpy as np
import matplotlib.pyplot as plt
import random

def approximatePi(numSamples):
    x, y, counter = 0, 0, 0
    for s in range(0, numSamples):
        x, y = random.uniform(-1, 1), random.uniform(-1,1)
        if (x**2 + y**2)<=1:
            counter += 1
    pi = 4* counter/numSamples
    return pi

errors = []
i = 10
while i <= 10000:
    p = approximatePi(i)
    error = round(np.abs(p-np.pi), 15)
    errors.append(error)
    i+=20
samples = np.linspace(10, 10000, len(errors))
plt.plot(samples, errors, 'r')
plt.show()
```
