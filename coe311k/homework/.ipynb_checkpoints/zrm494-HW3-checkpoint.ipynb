```python
import numpy as np
import matplotlib.pyplot as plt

def approx_sin(x, terms):
    running_sin = 0
    for i in range (0, terms):
        denom = 1
        for j in range (0, 2*i+1):
            denom = denom*(j+1) 
        running_sin += ((-1)**i) * (x**(2*i+1)) / denom
    return running_sin

x_vals = np.linspace(0, 10, 101)
one_vals = np.zeros(len(x_vals))
two_vals = np.zeros(len(x_vals))    
three_vals = np.zeros(len(x_vals))
four_vals = np.zeros(len(x_vals))
real_vals = np.zeros(len(x_vals))

for i in range (0, len(x_vals)):
   one_vals[i] = approx_sin(x_vals[i], 1) 
for i in range (0, len(x_vals)):
   two_vals[i] = approx_sin(x_vals[i], 2)
for i in range (0, len(x_vals)):
   three_vals[i] = approx_sin(x_vals[i], 3) 
for i in range (0, len(x_vals)):
   four_vals[i] = approx_sin(x_vals[i], 4) 
for i in range (0, len(x_vals)):
   real_vals[i] = np.sin(x_vals[i])

one_error = np.abs(one_vals - real_vals)
two_error = np.abs(two_vals - real_vals)
three_error = np.abs(three_vals - real_vals)
four_error = np.abs(four_vals - real_vals)

plt.subplot(2, 2, 1)
plt.plot(x_vals, one_error)
plt.xlabel('x')
plt.ylabel('one term error')

plt.subplot(2, 2, 2)
plt.plot(x_vals, two_error)
plt.xlabel('x')
plt.ylabel('two term error')

plt.subplot(2, 2, 3)
plt.plot(x_vals, three_error)
plt.xlabel('x')
plt.ylabel('three term error')

plt.subplot(2, 2, 4)
plt.plot(x_vals, four_error)
plt.xlabel('x')
plt.ylabel('four term error')

plt.tight_layout()
```

Truncation error is very dynamic. At smaller values of x (closer to zero, Maclaurin series centered at 0), the estimations are all significantly better than at larger values (farther from the Taylor origin). The effect becomes more dramatic with more terms, as with four terms an error of 1000 is reached at x=10 when the one term series only got to an error of 10. However, the four term series is able to stay much closer to the actual sin value for more x values. Creating a more accurate estimation for sin around a center makes truncation error siginificantly more extreme when farther away from the center.
