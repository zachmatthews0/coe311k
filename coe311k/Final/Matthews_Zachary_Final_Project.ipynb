```python
"""
With m = 360.7175 kg (¼ of car mass corresponding to only one wheel/one spring), a damping coefficient 
c = 2659.81 Ns/m, and spring constant k = 78450 N/m. The initial condition is y = 0.0762 corresponding 
to a typical speed bump of size 3 inches.

Damping ratio: 0.25
C = 2*(damping ratio) * sqrt(k*(m))
C = 2659.81 Ns/m
"""

# set initial conditions

M = 1442.87/4  # mass (kg)
C = 2659.81    # dampening coefficient (N*s/m)
K = 78450.0    # spring constant (N/m)

zetta = 0.25   # dampening ratio

Y_0 = -0.0762   # meters
V_0 = 0


step_sizes = [C/K+0.001, 0.01, 0.001, 0.0001]


def f(y, v):# second order
    return (-C*v-K*y)/M
```

```python
"""Description of scenario being modelled:

Scenario: A Volkswagen Taos hobbyist hits an average sized speedbump.
"""
pass
```

# Section 3: Implementation

```python
def analytical(t):
    w_n = math.sqrt(K/M)
    w_d = w_n * math.sqrt(1-zetta**2)
    B = (V_0 + zetta*w_n*Y_0)/w_d
    return math.exp(-zetta*w_n*t) * (Y_0*math.cos(w_d*t))
```

```python
def euler_forward(step):
    # Define parameters
    h = step  # step size
    t_start, t_end = 0, 1
    num_steps = int((t_end - t_start) / h)
    t_values = np.linspace(t_start, t_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    v_values = np.zeros(num_steps + 1)
    
    # Initial conditions
    y_values[0] = Y_0 
    v_values[0] = V_0
        
    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * v_values[i]
        v_values[i + 1] = v_values[i] + h * f(y_values[i], v_values[i])
    
    return t_values, y_values, v_values
```

```python
def rk4(step, t_end=5):
    # Define parameters
    t_start, t_end = 0, t_end
    num_steps = int((t_end - t_start) / step)
    t_values = np.linspace(t_start, t_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    v_values = np.zeros(num_steps + 1)

    # initial condition
    state = [Y_0, V_0]
    temp = [0, 0]
  
    # rk4 ks
    k = np.zeros((4, 2))
    
    for i in range(num_steps):
    
        k[0,0] = state[1]
        k[0,1] = f(state[0], state[1])

        temp = state + (0.5)*step*k[0,:]

        k[1,0] = temp[1]
        k[1,1] = f(state[0], state[1])

        temp = state + (0.5)*step*k[1,:]

        k[2,0] = temp[1]
        k[2,1] = f(state[0], state[1])

        temp = state + step*k[2,:]

        k[3,0] = temp[1]
        k[3,1] = f(state[0], state[1])

        # consolidate
        state = state + (step/6.) * (k[0,:]+2.*k[1,:]+2.*k[2,:]+k[3,:])
        
        y_values[i] = state[0]
        v_values[i] = state[1]
        
    return t_values, y_values, v_values
    
```

```python
"""Explanation

We implemented Euler's forward method by breaking down the 2nd order
equation into two first order ones, representing velocity and position
respectively. We then proceeded with simulating both over the duration.

We implemented RK4 using the same principle as above, however using 4
samples per timestep. Additionally, we kept the accelerations and 
velocities in one array, as multiplying each by the timestep (t) 
would yield the velocities and positions respectively. 

"""

pass
```

# Section 4: Solutions & Comparison

```python
# plot positions
ys = []
ws = []

for h in step_sizes:
    x_values, y_values, v_values = euler_forward(h)
    ys.append(y_values)
    ws.append(v_values)

    plt.plot(x_values, y_values, label=f"{h=}")
    
plt.xlabel('t')
plt.ylabel('y')
plt.title("Euler's Forward")
plt.legend()
plt.show()

# plot velocities
for h in step_sizes:
    x_values, y_values, v_values = euler_forward(h)

    plt.plot(x_values, v_values, label=f"{h=}")
    
plt.xlabel('t')
plt.ylabel('v')
plt.title("Euler's Forward")
plt.legend()
plt.show()


# phase portrait

for i in range(len(ys)):
    plt.plot(ys[i], ws[i], label=f"h={step_sizes[i]}")
    
plt.xlabel('y')
plt.ylabel('w')
plt.title("Phase Portraits")
plt.legend()
plt.show()
```

```python
ys = []
ws = []

# plot positions
for h in step_sizes:
    x_values, y_values, v_values = rk4(h)
    ys.append(y_values)
    ws.append(v_values)
    
    plt.plot(x_values, y_values, label=f"{h=}")
    
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.title("RK4")
plt.show()

# plot velocities
for h in step_sizes:
    x_values, y_values, v_values = rk4(h)

    plt.plot(x_values, v_values, label=f"{h=}")
    
plt.xlabel('t')
plt.ylabel('v')
plt.legend()
plt.title("RK4")
plt.show()


# phase portrait

for i in range(len(ys)):
    plt.plot(ys[i], ws[i], label=f"h={step_sizes[i]}")
    
plt.xlabel('y')
plt.ylabel('w')
plt.title("Phase Portriat")
plt.legend()
plt.show()
```

# Section 5: Stability Analysis

```python
# Stability analysis for euler's

for h in step_sizes:
    # coeff matrix
    
    # stability limit is C/K, which can be seen with the instability on the 
    # previous graphs when h is greater than that amount
    if (h < (C/K)):
        print(f"[{h=}]: Solution is stable")
    else:
        print(f"[{h=}]: Solution may not be entirely stable ;)")
```

```python
# Stability analysis for rk 4

for h in [2.8/math.sqrt(C/M) + 0.001, *step_sizes]:
    # coeff matrix
    
    # stability limit is 2.8 / sqrt(C/M)
    if (h < 2.8/math.sqrt(C/M)):
        print(f"[{h=}]: Solution is stable")
    else:
        print(f"[{h=}]: Solution may not be entirely stable ;)")
        
        

# set h to greater than stability limit
x_values, y_values, v_values = rk4(2.8/math.sqrt(C/M) + 0.001, 10)
ys.append(y_values)
ws.append(v_values)
    
plt.plot(x_values, y_values, label=f"{h=}")
    
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.title("RK4 instability")
plt.show()

```

```python
# Errors for euler
for h in step_sizes:
    x_values, y_values, v_values = euler_forward(h)
    y_reals = np.array([analytical(i) for i in x_values])

    plt.plot(x_values, y_values-y_reals, label=f"{h=}")
    
plt.xlabel('t')
plt.ylabel('error')
plt.title("Euler's Forward")
plt.legend()
plt.show()

# Errors for rk4
for h in step_sizes:
    x_values, y_values, v_values = rk4(h)
    y_reals = np.array([analytical(i) for i in x_values])

    plt.plot(x_values, y_values-y_reals, label=f"{h=}")
    
plt.xlabel('t')
plt.ylabel('error')
plt.title("RK4")
plt.legend()
plt.show()
```
