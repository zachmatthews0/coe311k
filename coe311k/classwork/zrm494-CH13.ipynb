```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt

def matrix_multiply(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
                
    return C

basic = np.zeros(8)
efficient = np.zeros(8)
x = np.arange(0, 8)

for w in range(0, 8):
    n = 3 * (2**w)
    x[w] = n
    A = np.zeros((n, n))
    for i in range(0, n):
        for j in range(0, n):
            A[i][j] = random.uniform(0, 10)
    B = np.zeros((n, n))
    for i in range(0, n):
        for j in range(0, n):
            B[i][j] = random.uniform(0, 10)
    print(n, 'x', n)
    
    basic1 = time.perf_counter()
    M = matrix_multiply(A, B)
    basic2 = time.perf_counter()
    basic[w] = basic2-basic1
    print('Basic time:', basic2-basic1)
    
    efficient1 = time.perf_counter()
    M = A @ B
    efficient2 = time.perf_counter()
    efficient[w] = efficient2-efficient1
    print('Efficient time:', efficient2-efficient1)
plt.plot(x, basic, label = 'naive')
plt.plot(x, efficient, label = 'optimized')
plt.legend()
plt.xlabel('n')
plt.ylabel('time')
```
