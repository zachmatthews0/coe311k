```python
import numpy as np
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    diag_dom = True
    for i in range (0, len(A)):
        row_sum = 0
        for j in range (0, len(A[0])):
            if (j != i):
                row_sum = row_sum + np.abs(A[i][j])
        if np.abs(A[i][i]) < row_sum:
            diag_dom = False
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k, diag_dom
        x = x_new
    return x, max_iterations, diag_dom

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    diag_dom = True
    for i in range (0, len(A)):
        row_sum = 0
        for j in range (0, len(A[0])):
            if (j != i):
                row_sum = row_sum + np.abs(A[i][j])
        if np.abs(A[i][i]) < row_sum:
            diag_dom = False
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k, diag_dom
        x = x_new
    return x, max_iterations, diag_dom
```

```python
A = np.array([

[0.582745, 0.48 , 0.10 , 0.  , 0.  ],
[0.48 , 1.044129, 0.46 , 0.10 , 0.  ],
[0.10 , 0.46 , 1.10431 , 0.44 , 0.10 ],
[0.  , 0.10 , 0.44 , 0.963889, 0.42 ],
[0.  , 0.  , 0.10 , 0.42 , 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)

x0 = np.zeros_like(b, dtype=float) 
max_iterations = 6000
tol = 1e-6

jsolution, jiterations, jdiag = jacobi(A, b, x0, tol, max_iterations)
gsolution, giterations, gdiag = gauss_seidel(A, b, x0, tol, max_iterations)

print(jsolution, jiterations, 'iterations, diagonally dominant', jdiag)
print(gsolution, giterations, 'iterations, diagonally dominant', gdiag)
```
