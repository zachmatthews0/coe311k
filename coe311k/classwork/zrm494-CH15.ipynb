```python
import numpy as np
import matplotlib.pyplot as plt

# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y
```

```python
h_high = np.linspace(0.05, 1, 20)

# Define base parameters
h = 0.01  # step size
x_start, x_end = 0, 2
num_steps = int((x_end - x_start) / h)
x_values = np.linspace(x_start, x_end, num_steps + 1)
y_values = np.zeros(num_steps + 1)

# Initial condition
y_values[0] = 1  # y(0) = 1

# Implement Euler's method
for i in range(num_steps):
    y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
    
# Plot the results
plt.plot(x_values, y_values, label="Euler's Method Approximation")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Approximation of dy/dx = -2y using Euler's Method")
plt.legend()
plt.show()
    
for i in range(0, len(h_high)):
    h = np.round(0.05*(i + 1), 2)
    low_num_steps = int((x_end - x_start) / h)
    worse_x_values = np.linspace(x_start, x_end, low_num_steps + 1)
    worse_y_values = np.zeros(low_num_steps + 1)
    worse_y_values[0] = 1  # y(0) = 1
    
    for i in range(low_num_steps):
        worse_y_values[i + 1] = worse_y_values[i] + h * f(worse_x_values[i], worse_y_values[i])
    errors = np.zeros(len(worse_x_values))
    
    for i in range(0, len(errors)):
        errors[i] = np.abs(worse_y_values[i] - y_values[i*int((len(x_values)/len(worse_x_values)))])
    plt.plot(x_values, y_values, label="Euler's Method Approximation")
    plt.plot(worse_x_values, worse_y_values, label="Higher step size Euler's Method Approximation")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Approximation of dy/dx = -2y using Euler's Method Against Higher Step Size")
    plt.legend()
    plt.show()
    print('h = ', h)
    total_error = np.sum(errors)
    print('Pointwise errors: ', errors)
    print('Accumulated error: ', total_error)
```
