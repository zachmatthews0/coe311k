```python
def f1(x, y):
    return x*x + y*y

x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)

X, Y = np.meshgrid(x, y)
Z = f1(x, y)
```

```python
X
```

```python
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.countour3D(X, Y, Z, 50, cmap='hot')
ax.set_xlabel('x')
ax.view_init(45, 0)
```

```python
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='hot')
ax.set_xlabel('x')
ax.view_init(45, 0)
```
