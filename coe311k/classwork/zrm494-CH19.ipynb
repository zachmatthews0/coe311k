```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, y):
    return (-1000 * y + 3000 - 2000*np.exp(-t))

def exact(t):
    return 3 - 0.998*np.exp(-1000*t) - 2.002*np.exp(-t)
    
# Define parameters
h_list = np.array([.1, .01, .001, .00001])# step size
t_start, t_end = 0, 30

t_base = np.linspace(0, 30, 301)
y_analytic = np.zeros(len(t_base))

for i in range (len(t_base)):
    y_analytic[i] = exact(t_base[i])

for h in h_list:
    num_steps = int((t_end - t_start) / h)
    t_values = np.linspace(t_start, t_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1 # y(0) = 0

    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(t_values[i], y_values[i])
    
    # Plot the results
    plt.plot(t_values, y_values, label="Euler's Method Approximation")
    plt.plot(t_base, y_analytic, label="Analytics Values")
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title("Approximation of Stiff Equation using Euler's Method, h = " + str(h), pad=20)
    plt.legend()
    plt.show()
```
