```python
import numpy as np

def piapprox(terms):
    sum = 0;
    for i in range (0, terms):
        sum = sum + ((-1)**i/(2*i+1))                
    return sum*4
                     
for i in range (0, int(5000/100)):
    pi = piapprox(i*100 + 1)
    error = np.abs(pi-np.pi)
    print(error)
```
