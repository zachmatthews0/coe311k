```python
import numpy as np
import matplotlib.pyplot as plt
import random

def approximatePi(numSamples):
    x, y, counter = 0, 0, 0
    for s in range(0, numSamples):
        x, y = random.uniform(-1, 1), random.uniform(-1,1)
        if (x**2 + y**2)<=1:
            counter += 1
    pi = 4* counter/numSamples
    return pi

errors = []
errors_rel = []
for i in range (0,100):
    p = approximatePi(1000)
    error = round(np.abs(p-np.pi), 15)
    errors.append(error)
    errors_rel.append(error/np.pi)
samples = np.linspace(1, 100, 100)
print(samples)
plt.subplot(1, 2, 1)
plt.plot(samples, errors, 'r')
plt.subplot(1, 2, 2)
plt.plot(samples, errors_rel, 'b')
plt.show()
```
