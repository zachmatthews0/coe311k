```python
import numpy as np
import matplotlib.pyplot as plt
# Define the function and the domain
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi # Interval
N = 50 # Number of sample points
# Create a uniform grid
x = np.linspace(a, b, N)
y = f(x)

# Plot the sampled points
plt.plot(x, y, 'o-', label='Sampled Points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Uniform Sampling of f(x) = sin(x)')
plt.legend()
plt.show()
```

```python
# Step size
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
df_dx_c = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation for the derivative
df_dx_f = (f(x + h) - f(x)) / (h)

# Central Difference Approximation for the derivative
df_dx_b = (f(x) - f(x - h)) / (h)

# Plot the approximation and the true derivative
plt.plot(x, np.cos(x), label='Exact Derivative')
plt.plot(x, df_dx_c, '--', label='Central Difference Approximation')
plt.plot(x, df_dx_f, '--', label='Forward Difference Approximation')
plt.plot(x, df_dx_b, '--', label='Backward Difference Approximation')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Approximating Derivative off(x) = sin(x) using Central Difference")
plt.legend()
plt.show()
```

Forward difference seems to "get ahead" of the curve - goes down before the exact derivative and goes up before the exact derivative. Backward difference is the opposite - it is usually slightly behind. Central difference is a very good approximation and is nearly indistinguishable from the exact derivative.
