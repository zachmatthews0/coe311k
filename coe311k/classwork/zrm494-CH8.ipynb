```python
import numpy as np

def diag_dominance(A):
    diag_dom = True
    for i in range (0, len(A)):
        row_sum = 0
        for j in range (0, len(A[0])):
            if (j != i):
                row_sum = row_sum + np.abs(A[i][j])
        if np.abs(A[i][i]) < row_sum:
            diag_dom = False
    return diag_dom

A = ([[1, 2, 1], [2, 3, 2], [4, 2, 1]])
print(A, 'Is the matrix diagonally dominant?', diag_dominance(A))

A = ([[4, 2, 1], [6, 3, 3], [4, 2, 3]])
print(A, 'Is the matrix diagonally dominant?', diag_dominance(A))

A = ([[4, 2, 1], [3, 5, 2], [2, 1, 4]])
print(A, 'Is the matrix diagonally dominant?', diag_dominance(A))
```
