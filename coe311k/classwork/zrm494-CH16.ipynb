```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
h = 0.1 # step size
x_start, x_end = 0, 10
num_steps = int((x_end - x_start) / h)
x_values = np.linspace(x_start, x_end, num_steps + 1)
y_values = np.zeros(num_steps + 1) 

# Initial condition
y_values[0] = 1 # y(0) = 1

def exact(x, y0, lam):
    return y0*(np.exp(x*lam))

# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y

# Implement Euler's method
for i in range(num_steps):
    y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])

local_errors = np.zeros(len(x_values))
global_errors = np.zeros(len(x_values))

for i in range(len(x_values)):
    local_errors[i] = np.abs(y_values[i]-exact(x_values[i], y_values[0], -2))
global_errors[0] = local_errors[0]
for i in range(len(x_values) - 1):
    global_errors[i+1] = local_errors[i+1] + global_errors[i]
    
# Plot the results
plt.plot(x_values, y_values, label="Euler's Method Approximation: h = 0.1")
plt.plot(x_values, local_errors, label="Local Error")
plt.plot(x_values, global_errors, label="Global Error")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Approximation of dy/dx = -2y using Euler's Method: h = 0.1")
plt.legend()
plt.show()
```
