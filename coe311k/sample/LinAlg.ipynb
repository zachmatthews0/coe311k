```python
B = np.array([[1, 2, 3], [4, 5, 5], [7, 8, 9]])
```

```python
e_value, e_vector = np.linalg.eig(B)
print(e_value)
```

```python
print(e_vector)
```

```python
A_inv = np.linalg.inv(B)
print(A_inv)
```

```python
print(np.dot(B, A_inv))
```
